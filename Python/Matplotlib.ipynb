## Line Charts
Line charts are created using the plot() method.

Line charts are a good choice for showing trends, 

```python

years = [2006, 2008, 2010, 2012, 2014, 2016, 2018]
sales = [500.6, 543.3, 654.9, 600.3, 766.6, 889.7, 848.3]

fig = plt.figure()  # an empty figure with no axes

fig, axes =  plt.subplots(1, 1)

axes.set_ylabel("Millions of $")
axes.set_xlabel("Year")
axes.set_title("Money Spent on Video Games")

axes.plot(years, sales, color='green', marker='o', linestyle='solid')

plt.show()
```

```python

plt.plot(years, sales, color='green', marker='o', linestyle='solid')
plt.title("Money Spent on Video Games")
plt.xlabel("Year")
plt.ylabel("Millions of $")
plt.show()
```

```python
fig, ax1 = plt.subplots()

t = [1,2,3,4,5,6,7,8,9]
data1     = [1, 2, 4, 8, 16, 32, 64, 128, 256]
data2 = [256, 128, 64, 32, 16, 8, 4, 2, 1]

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

## Bar Charts

A bar chart is a good choice when you want to show how some quantity varies among
some discrete set of items. 

```python
genre = ["Action", "Comedy", "Sci-Fi", "Fantasy", "Romance"]
num_oscars = [20, 32, 22, 28, 14]

#Center bars
plt.xticks([i + 0.5 for i, _ in enumerate(genre)], genre)
xs = [i + 0.5 for i, _ in enumerate(genre)]
plt.bar(xs, num_oscars, color = 'blue')
plt.xlabel("Genre")
plt.ylabel("People")
plt.title("Favorite Movie Genres")
# label x-axis with movie names at bar centers

plt.show()
```

```python
fig = plt.figure()

fig, axes =  plt.subplots(1, 1)


axes.set_xticks([i + 0.5 for i, _ in enumerate(genre)], genre)
xs = [i + 0.5 for i, _ in enumerate(genre)]
axes.bar(xs, num_oscars, color = 'blue')
axes.set_xlabel("Genre")
axes.set_ylabel("People")
axes.set_title("Favorite Movie Genres")
# label x-axis with movie names at bar centers
plt.show()
```

A bar chart is used for plotting histograms of bucketed numeric
values. This enables you to visually explore how the values are distributed

```python
from collections import Counter
heights = [44,49,32,51,35,29,41,61,53,81,36,43,27,46]
decile = lambda heights: heights // 10 * 10 # round down to nearest 10 
histogram = Counter(decile(height) for height in heights)


# x, height(s) of bars, width
plt.bar([x - 4 for x in histogram.keys()], # shift each bar to the left by 4
        histogram.values(),                
        8, align='edge')                                 # give each bar a width of 8

#x-axis from -5 to 105, y-axis from 0 to 5
plt.axis([-5, 105, 0, 6]) 

# x-axis labels - 0, 10, ..., 90, 100 
plt.xticks([10 * i for i in range(11)])    
plt.xlabel("Decile (Feet)")
plt.ylabel("# of Trees")
plt.title("Tree Heights")
plt.show()
```
