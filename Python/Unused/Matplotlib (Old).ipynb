A bar chart is used for plotting histograms of bucketed numeric
values. This enables you to visually explore how the values are distributed

````python
from collections import Counter
heights = [44,49,32,51,35,29,41,61,53,81,36,43,27,46]
decile = lambda heights: heights // 10 * 10 # round down to nearest 10 
histogram = Counter(decile(height) for height in heights)


# x, height(s) of bars, width
plt.bar([x - 4 for x in histogram.keys()], # shift each bar to the left by 4
        histogram.values(),                
        8, align='edge')                                 # give each bar a width of 8

#x-axis from -5 to 105, y-axis from 0 to 5
plt.axis([-5, 105, 0, 6]) 

# x-axis labels - 0, 10, ..., 90, 100 
plt.xticks([10 * i for i in range(11)])    
plt.xlabel("Decile (Feet)")
plt.ylabel("# of Trees")
plt.title("Tree Heights")
plt.show()
````

## Scatterplot

A scatterplot can be used for visualizing the relationship between two paired sets of data.

````python
weight = [174, 155, 189, 165, 151, 162, 140, 180, 190]
height = [ 60,  48,  66,  59,  53,  50,  45,  64,  67]
names =  ['Bill', 'Jane', 'Joe', 'Mike', 'Kelly', 'Marv', 'Tom', 'Tim', 'Janet']
plt.scatter(weight, height)
# label each point
for label, w_count, h_count in zip(names, weight, height):
    plt.annotate(label,
        xy=(w_count, h_count), # put the label with its point
        xytext=(5, -5),                  # but slightly offset
        textcoords='offset points')
plt.title("Weight vs. Height")
plt.xlabel("Weight (lb)")
plt.ylabel("Height (in)")
plt.show()
````

Including a call to plt.axis("equal") might show a more accurate picture with a better fit. Letting matplotlib automatically set the scale, might provide a misleading picture

````python
weight = [174, 155, 189, 165, 151, 162, 140, 180, 190]
height = [ 60,  48,  66,  59,  53,  50,  45,  64,  67]
names =  ['Bill', 'Jane', 'Joe', 'Mike', 'Kelly', 'Marv', 'Tom', 'Tim', 'Janet']
plt.scatter(weight, height)
# label each point
for label, w_count, h_count in zip(names, weight, height):
    plt.annotate(label,
        xy=(w_count, h_count), # put the label with its point
        xytext=(5, -5),                  # but slightly offset
        textcoords='offset points')
plt.title("Weight vs. Height")
plt.xlabel("Weight (lb)")
plt.ylabel("Height (in)")
plt.axis("equal")
plt.show()
````

## Creating a Basic Figure

To create a graph, you first declare a figure object by assigning it to a variable:

```Python
fig = plt.figure()
```
A figure will display in Jupyter Notebook without calling this but it is still good practice to declare it.

We can then call a function that will plot data onto the figure. The way the data is represented depends on the function called. In this example,, we'll be using `plt.plot()` which is a means to plot data as a y verses x graph.

The first two parameters take the X-axis and Y-axis values, respectively. They can be scalar values or lists.

```Python
plt.plot(x,y)
```

The size of the values of given to both `x` and `y` must be the same or an error is thrown.

Finally, to display the figure, run:

```Python
plt.show()
```

This is not required in Jupyter Notebook to display the figure but it is good practice to use it anyway. Doing so will prevent an object reference message from displaying (e.g. [<matplotlib.lines.Line2D at 0x00000000000>] )

````python
x = [1,3,5,7,8]
y = [3,7,8,13,11]

fig = plt.figure()

plt.plot(x, y)

plt.show()
````

The result is a line graph connecting at the points passed to the function.

However, the way the figure is displayed may vary depending on the default properties Jupyter Notebook sets.

As a result, it is better to manually set the properties of the figure.

Here are some functions that can change the properties of figures:

### plt.title()

```Python
plt.title(label)
```

Sets the title of the figure where `label` is a string used for the title

````python
fig = plt.figure()

plt.plot(x, y)

plt.title("Graph Title")

plt.show()
````

### plt.xlabel() and plt.ylabel

```Python
plt.xlabel(xlabel)
plt.ylabel(ylabel)
```

Sets the label for the x-axis and y-axis of the figure where `xlabel` and `ylabel` is a string used for the axes' respective labels. 

````python
fig = plt.figure()

plt.plot(x, y)

plt.title("Graph Title")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
````

### plt.xlim() and plt.ylim()

Get or set the x-limits and y-limits of the current axes.

By default, autoscaling sets the limits of the axes based on the maximum and minimum values of the data provided.

Setting limits turns autoscaling off for the respective axes.

There are several ways to set the maximum and minimum values:

1) Individual arguments

```Python
plt.xlim(bottom, top)
plt.ylim(bottom, top)
```

Where `bottom` is the minimum axis value and `top` is the maximum axis value.

2) Arguments in a tuple

```Python
plt.xlim((bottom, top))
plt.ylim((bottom, top))
```

3) As keyword arguments

```Python
plt.xlim(bottom=val)
plt.xlim(top=val)

plt.ylim(bottom=val)
plt.ylim(top=val)
```

This method lets you set the minimum and maximum values sepearately without the need to set both.


You can also get the limit values and assign them to variables by calling:

```Python
x_bottom, x_top = xlim()
y_bottom, y_top = ylim()
```

Here is a graph that starts at the origin

````python
fig = plt.figure()

plt.plot(x, y)

plt.title("Graph Title")
plt.xlabel("X axis")
plt.ylabel("Y axis")


plt.xlim(0, 14)
plt.ylim(0, 14)

plt.show()
````

Here is a graph that includes negative values

````python
fig = plt.figure()

plt.plot(x, y)

plt.title("Graph Title")
plt.xlabel("X axis")
plt.ylabel("Y axis")


plt.xlim(-10, 10)
plt.ylim(0, 15)

plt.show()
````

### Line Charts (plot())

As already demonstrated in the previous section, line charts are created using the `plot()` method.

The first two parameters take the X-axis and Y-axis values, respectively. They can be scalar values or lists.

```Python
plt.plot(x,y)
```

The size of the values of given to both `x` and `y` must be the same or an error is thrown.

Line charts are a good choice for showing trends.

For example, if we wanted to see how well the sales of video games over a period of 12 years, sa line graph might look like:

````python
years = [2006, 2008, 2010, 2012, 2014, 2016, 2018]
sales = [500.6, 543.3, 654.9, 600.3, 766.6, 889.7, 848.3]

fig = plt.figure()


plt.ylabel("Millions of $")
plt.xlabel("Year")
plt.title("Money Spent on Video Games")

plt.plot(years, sales)

plt.show()
````

#### fmt Parameter

```Python
plt.plot(x,y,fmt)
```

A third paramter can be added to the function to indicate the style.

The string 'go-' will make the line green, data points will be indicated as circles, and the line will be solid 



````python
fig = plt.figure()

plt.ylabel("Millions of $")
plt.xlabel("Year")
plt.title("Money Spent on Video Games")

plt.plot(years, sales, 'go-')

plt.show()
````

The same result can be achieved by passing Line2D properties as arguments

````python
fig = plt.figure()

plt.ylabel("Millions of $")
plt.xlabel("Year")
plt.title("Money Spent on Video Games")

plt.plot(years, sales, color='green', marker='o', linestyle='solid')

plt.show()
````

The linestyle can be removed also

````python
fig = plt.figure()

plt.ylabel("Millions of $")
plt.xlabel("Year")
plt.title("Money Spent on Video Games")

plt.plot(years, sales, 'go')

plt.show()
````

#### Plotting Multiple Datasets

Multiple datasets can be passed into the `plot()` function along with it's formatting.

```Python
plt.plot(x1,y1,fmt1,x2,y2,fmt2,...)
```
Building on our previous example, we will compare the trends of sales for two video game genres and their total.

````python
years = [2006, 2008, 2010, 2012, 2014, 2016, 2018]
total_sales = [500.6, 543.3, 654.9, 600.3, 766.6, 889.7, 848.3]
adventure_sales = [234.1, 291.2, 377.4, 302.2, 405.1, 462.4, 422.3]
action_sales = [266.5, 252.1, 277.5, 298.1, 361.5, 427.3, 426.0]

fig = plt.figure()

plt.ylabel("Millions of $")
plt.xlabel("Year")
plt.title("Money Spent on Video Games")

plt.plot(years, total_sales, 'go-', years, adventure_sales, 'ro--', years, action_sales, 'bo--')

plt.show()
````
