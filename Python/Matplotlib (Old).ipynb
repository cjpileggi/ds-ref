## Line Charts
Line charts are created using the plot() method.

Line charts are a good choice for showing trends, 

```python

years = [2006, 2008, 2010, 2012, 2014, 2016, 2018]
sales = [500.6, 543.3, 654.9, 600.3, 766.6, 889.7, 848.3]

fig = plt.figure()  # an empty figure with no axes

fig, axes =  plt.subplots(1, 1)

axes.set_ylabel("Millions of $")
axes.set_xlabel("Year")
axes.set_title("Money Spent on Video Games")

axes.plot(years, sales, color='green', marker='o', linestyle='solid')

plt.show()
```

Multiple Lines and Multiple Y axis can be plotted on the same graph

```python
fig, ax1 = plt.subplots()

month = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep', 'Oct', 'Nov', 'Dec']
temp     = [2, -2, 5, 11, 18, 22, 26, 29, 25, 19, 11, 7]
rain = [22, 19, 25, 36, 30, 19, 10, 5, 12, 19, 16, 20]

color = 'tab:red'
ax1.set_xlabel('month')
ax1.set_ylabel('Avg. Temperature (C)', color=color)
ax1.plot(month, temp, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Avg. Rainfall (mm)', color=color)  # we already handled the x-label with ax1
ax2.plot(month, rain, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

## Bar Charts

A bar chart is a good choice when you want to show how some quantity varies among
some discrete set of items. 

```python
genre = ["Action", "Comedy", "Sci-Fi", "Fantasy", "Romance"]
num_oscars = [20, 32, 22, 28, 14]

#Center bars
plt.xticks([i + 0.5 for i, _ in enumerate(genre)], genre)
xs = [i + 0.5 for i, _ in enumerate(genre)]
plt.bar(xs, num_oscars, color = 'blue')
plt.xlabel("Genre")
plt.ylabel("People")
plt.title("Favorite Movie Genres")
# label x-axis with movie names at bar centers

plt.show()
```

```python
fig = plt.figure()

fig, axes =  plt.subplots(1, 1)


axes.set_xticks([i + 0.5 for i, _ in enumerate(genre)], genre)
xs = [i + 0.5 for i, _ in enumerate(genre)]
axes.bar(xs, num_oscars, color = 'blue')
axes.set_xlabel("Genre")
axes.set_ylabel("People")
axes.set_title("Favorite Movie Genres")
# label x-axis with movie names at bar centers
plt.show()
```

A bar chart is used for plotting histograms of bucketed numeric
values. This enables you to visually explore how the values are distributed

```python
from collections import Counter
heights = [44,49,32,51,35,29,41,61,53,81,36,43,27,46]
decile = lambda heights: heights // 10 * 10 # round down to nearest 10 
histogram = Counter(decile(height) for height in heights)


# x, height(s) of bars, width
plt.bar([x - 4 for x in histogram.keys()], # shift each bar to the left by 4
        histogram.values(),                
        8, align='edge')                                 # give each bar a width of 8

#x-axis from -5 to 105, y-axis from 0 to 5
plt.axis([-5, 105, 0, 6]) 

# x-axis labels - 0, 10, ..., 90, 100 
plt.xticks([10 * i for i in range(11)])    
plt.xlabel("Decile (Feet)")
plt.ylabel("# of Trees")
plt.title("Tree Heights")
plt.show()
```

## Scatterplot

A scatterplot can be used for visualizing the relationship between two paired sets of data.

```python
weight = [174, 155, 189, 165, 151, 162, 140, 180, 190]
height = [ 60,  48,  66,  59,  53,  50,  45,  64,  67]
names =  ['Bill', 'Jane', 'Joe', 'Mike', 'Kelly', 'Marv', 'Tom', 'Tim', 'Janet']
plt.scatter(weight, height)
# label each point
for label, w_count, h_count in zip(names, weight, height):
    plt.annotate(label,
        xy=(w_count, h_count), # put the label with its point
        xytext=(5, -5),                  # but slightly offset
        textcoords='offset points')
plt.title("Weight vs. Height")
plt.xlabel("Weight (lb)")
plt.ylabel("Height (in)")
plt.show()
```

Including a call to plt.axis("equal") might show a more accurate picture with a better fit. Letting matplotlib automatically set the scale, might provide a misleading picture

```python
weight = [174, 155, 189, 165, 151, 162, 140, 180, 190]
height = [ 60,  48,  66,  59,  53,  50,  45,  64,  67]
names =  ['Bill', 'Jane', 'Joe', 'Mike', 'Kelly', 'Marv', 'Tom', 'Tim', 'Janet']
plt.scatter(weight, height)
# label each point
for label, w_count, h_count in zip(names, weight, height):
    plt.annotate(label,
        xy=(w_count, h_count), # put the label with its point
        xytext=(5, -5),                  # but slightly offset
        textcoords='offset points')
plt.title("Weight vs. Height")
plt.xlabel("Weight (lb)")
plt.ylabel("Height (in)")
plt.axis("equal")
plt.show()
```
